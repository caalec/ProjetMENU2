### Calcul du champ focalisé

En appliquant une nouvelle fois le principe de réciprocité et en utilisant comme signal d'entrée les signaux retournés obtenus juste avant, on parvient à focaliser le champ sur la source initiale.



Malheureusement, comme nous avons eu un problème de décalage temporel dès l'obtention de la pression (comme expliqué), notre tache focale n'est pas au bon endroit. Cette tache devrait être positionnée en $r=r_s$ et $z=z_s$ soit $t=z/c=10/1.5=6.67s$. Nous ne sommes pas parvenus à régler ce problème.

La tache focale se situe approximativement en $r=12.5mm$ et $t=12s$.

### Forme de la tâche focale

Théoriquement, on peut la tâche focale a parallèlement au réseau de trasducteurs une taille proportionnelle à 
$$\lambda \frac{F}{O}$$
où
- $\lambda$ est la longueur d'onde
- $F$ est la distance focale (profondeur de la source initiale)
- $O$ est l'ouverture de la surface diffractante (ouverture totale du réseau de transducteurs

On pourrait vérifier cette propriété en faisant varier un des paramètres tout en gardant les deux autres constants, 
et ce pour chacun des paramètres.
On fait varier F en faisant varier z_source.
On fait varier O en faisant varier a ou N_r.
On fait varier lambda en faisant varier c ou f_centrale.
La taille de la tache focale peut être obtenue en regardant où la valeur absolue du champ de pression dépasse 
un certain seuil qu'on fixera.
